```python
from qiskit import *
from math import pi
import numpy as np
from qiskit.visualization import plot_bloch_multivector, plot_histogram, array_to_latex
```

```python
qc= QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
display(qc.draw(output='mpl'))
#JupyterNotebook에서display는 print와 비슷하나 많은 내용을 출력해야 할 때 쓰임
simulator = Aer.get_backend('aer_simulator')
qc.save_statevector()
final_state = execute(qc, simulator, shots=1024).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Statevector = }"))
plot_bloch_multivector(final_state)
```

```python
#hadamard gate는 다시 대입하게 되면 원래대로 되돌려줌
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)
display(qc.draw(output='mpl'))

qc.save_unitary()
unitary = execute(qc, simulator, shots=1024).result().get_unitary()
array_to_latex(unitary, prefix="\\text{Circuit =}\n")
```

Phase Kickback
Quantum Algorithm에서 중요한 개념
Kickback : qubit에 gate를 통해 eigenvalue가 더해진 상태(Kicked back)를 controlled operation을 통해  다른 qubit으로 만들어주는 것


```python
#T-gate에서의 Kickback
qc = QuantumCircuit(2)
#controlled p-gate : pgate에서 pi/4: tgate가 됨
qc.cp(pi/4, 0, 1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(2)
qc.cp(pi/4, 0, 1)
display(qc.draw(output='mpl'))
qc.save_unitary()
unitary = execute(qc, simulator, shots=1024).result().get_unitary()
array_to_latex(unitary, prefix="\\text{Controlled-T} =\n")
```
