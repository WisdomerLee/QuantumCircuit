```python
from qiskit import *
from math import pi
import numpy as np
from qiskit.visualization import plot_bloch_multivector, plot_histogram, array_to_latex
```

```python
qc= QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
display(qc.draw(output='mpl'))
#JupyterNotebookdisplay print       
simulator = Aer.get_backend('aer_simulator')
qc.save_statevector()
final_state = execute(qc, simulator, shots=1024).result().get_statevector()
display(array_to_latex(final_state, prefix="\\text{Statevector = }"))
plot_bloch_multivector(final_state)
```

```python
#hadamard gate     
qc = QuantumCircuit(2)
qc.h(0)
qc.h(1)
qc.cx(0,1)
qc.h(0)
qc.h(1)
display(qc.draw(output='mpl'))

qc.save_unitary()
unitary = execute(qc, simulator, shots=1024).result().get_unitary()
array_to_latex(unitary, prefix="\\text{Circuit =}\n")
```

Phase Kickback
Quantum Algorithm  
Kickback : qubit gate  eigenvalue  (Kicked back) controlled operation    qubit  


```python
#T-gate Kickback
qc = QuantumCircuit(2)
#controlled p-gate : pgate pi/4: tgate 
qc.cp(pi/4, 0, 1)
qc.draw(output='mpl')
```

```python
qc = QuantumCircuit(2)
qc.cp(pi/4, 0, 1)
display(qc.draw(output='mpl'))
qc.save_unitary()
unitary = execute(qc, simulator, shots=1024).result().get_unitary()
array_to_latex(unitary, prefix="\\text{Controlled-T} =\n")
```
